# Random Forest – Giải thủ công

Bài toán dự đoán *Tình trạng da* bằng Random Forest (giải thích thủ công).

## 1. Dữ liệu ban đầu

```
import pandas as pd

data = {
    'STT': [1,2,3,4,5,6,7],
    'Màu da': ['Sẫm','Sẫm','Sáng','Ngăm','Ngăm','Sáng','Nâu'],
    'Tầm vóc': ['Trung bình','Cao','Thấp','Thấp','Trung bình','Trung bình','Trung bình'],
    'Thể trạng': ['Nhẹ','Trung bình','Trung bình','Trung bình','Nặng','Nặng','Nhẹ'],
    'Dùng kem': ['Không','Có','Có','Không','Không','Có','Có'],
    'Tình trạng da': ['Cháy nắng','Không','Không','Cháy nắng','Cháy nắng','Không','?']
}

df = pd.DataFrame(data)
df
```

## 2. Ý tưởng Random Forest

- Random Forest gồm nhiều Decision Tree
- Mỗi cây dùng một tập thuộc tính khác nhau
- Kết quả cuối cùng là **bỏ phiếu đa số**

## 3. Cây quyết định 1 (Dùng kem chống nắng)

Luật:
- Nếu Dùng kem = Có → Không
- Nếu Dùng kem = Không → Cháy nắng

```
prediction_tree1 = "Không"  # mẫu 7 dùng kem = Có
prediction_tree1
```

## 4. Cây quyết định 2 (Màu da)

Luật:
- Sáng → Không
- Sẫm / Ngăm / Nâu → Cháy nắng

```
prediction_tree2 = "Cháy nắng"  # màu da = Nâu
prediction_tree2
```

## 5. Cây quyết định 3 (Thể trạng)

Luật:
- Nhẹ → Cháy nắng
- Trung bình / Nặng → Không

```
prediction_tree3 = "Cháy nắng"  # thể trạng = Nhẹ
prediction_tree3
```

## 6. Bỏ phiếu (Majority Voting)

```
votes = [prediction_tree1, prediction_tree2, prediction_tree3]
votes
```

```
from collections import Counter
Counter(votes)
```

## 7. Kết luận

Đa số cây dự đoán **Cháy nắng** → Đây là kết quả cuối cùng cho mẫu số 7.
