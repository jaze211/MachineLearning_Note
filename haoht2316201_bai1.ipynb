```python
A = np.random.randint(0, 10,(5, 3))
print(A)
print("Trung bình (mean):", np.mean(A))
print("Trung vị (median):", np.median(A))
print("Phương sai (variance):", np.var(A))
print("Độ lệch chuẩn (std):", np.std(A))

p25 = np.percentile(A, 25)
p50 = np.percentile(A, 50)
p75 = np.percentile(A, 75)

print("P25:", p25)
print("P50 (median):", p50)
print("P75:", p75)
```
